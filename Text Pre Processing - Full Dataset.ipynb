```python
#Try to implement into Hadith content (Full Dataset)

#2-Read data into pandas dataframe
df = pd.read_csv(r"C:\Users\Acer\anaconda3\activity\Sahih Bukhari Dataset.csv")
df
```

```python
#3-Tokenization (split into tokens)

def word_tokenization(text):
    text = word_tokenize(text.lower())
    return " ".join([word for word in str(text).split()])

df["Text Tokenization"] = df["Text"].apply(lambda text: word_tokenization(text))
df
```

```python
#4-Removing punctuations 
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text

df["Text Remove Punc"] = df['Text Tokenization'].apply(remove_punctuations)
df
```

```python
#5-Removing stopwords 
stops = set(final_stopword_list)
def remove_stopwords(text):
    return " ".join([word for word in str(text).split()
                    if word not in stops])

df["Text Remove Stops"] = df["Text Remove Punc"].apply(lambda text: remove_stopwords(text))
df
```

```python
#6-Stemming (valid stems)
#Example:
#heard allah apostle saying reward deeds depends upon intentions every person get reward according intended 
#whoever emigrated worldly benefits woman marry emigration emigrated

#heard allah apostl say reward deed depend upon intent everi person get reward accord intend 
#whoever emigr worldli benefit woman marri emigr emigr 

porter = PorterStemmer()
def stemming(text):
    token_words=word_tokenize(text)
    token_words
    stem_sentence=[]
    for word in token_words:
        stem_sentence.append(porter.stem(word))
        stem_sentence.append(" ")
    return "".join(stem_sentence)

df["Text Cleaned"] = df["Text Remove Stops"].apply(lambda text: stemming(text))  
#df = df.drop(columns=['unstemmed']) # Get rid of the unstemmed column.
df 
```
